# Contour Plot

## Import modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Dataset

```python
n = np.linspace(-3.0, 3.0, 100)

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

## Contour Plot

- contour plot digunakan menggmabarkan data 3 dimensi dalam bentuk 2 dimensi

code untuk memebuat contour plot

```python
plt.contour(x, y, z)
plt.colorbar()

plt.title('Contour Plot')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

## Filled Contour Plot

code untuk memebuat contour plot berwarna penuh (full)

```python
plt.contourf(x, y, z)
plt.colorbar()

plt.title('Filled Contour Plot')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```
