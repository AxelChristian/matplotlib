## Predefined Styles

```python
plt.style.available
```

code untuk menggunakan predefined style

```python
plt.style.use('seaborn-pastel')
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='linear')
plt.plot(x, np.log(x), 'g^-', label='logarithmic')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot')

plt.show()
```

##  Buat Style Sendiri

code untuk membuat style sendiri :

- buat file.txt
    - masukan konfigurasi yang ingin diubah misal :
            - axes.titlesize : 24
            - axes.labelsize : 20
            - lines.linewidth : 3
            - lines.markersize : 10
            - xtick.labelsize : 16
            - ytick.labelsize : 16
- save file dengan nama tertentu
- import file yang dibuat

```python
plt.style.use('./style_ku.mplstyle')
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot')

plt.show()
```

##  Multiple style

code untuk mengimplementasikan multiple style (lebih dari 1), hanya berlaku bagi style yang bersifat overlap

```python
plt.style.use(['classic', 'ggplot'])
```

```python
x = np.linspace(1, 10, 15)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')

plt.legend()
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Plot')

plt.show()
```

## Temporary styling

```python
plt.style.use('bmh')

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

code untuk membuat plot memiliki dark background hanya pada cell tersebut

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')

plt.show()
```
