```python
#Multi plot dalam satu fungsi plot()
```

```python
#Cara panjang
t=np.arange(0., 5., 0.2) #buat array urutan (nilai start, nilai akhir, interval)(0.= float), 
                         #5 tidak masuk karena sebagai pembatas
t 
```

```python
plt.plot(t, t, "r--") #r = merah, -- = putus putus
plt.plot(t, t**2, "bs") #b = biru, s= square
plt.plot(t, t**3, "g^") #g = hijau, ^ = segitiga
plt.show()
```

```python
#fungsi plot()
plt.plot(t, t, "r--", t, t**2, "bs", t, t**3, "g^")
plt.show()
```

```python
#Ploting dengan keywords
```

```python
data = {
    "a": np.arange(50),
    "c": np.random.randint(0, 50, 50),
    "d": np.random.randn(50)
}
data["b"] = data["a"] + 10 * np.random.randn(50)
data["d"] = np.abs(data["d"]) * 100

data
```

```python
plt.scatter("a","b",data=data) #data yang ada pada a dan b
plt.xlabel("entry a") 
plt.ylabel("entry b")
plt.show()
```

```python
plt.scatter("a", "b", c="c", s="d", data=data)
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.show()
```
