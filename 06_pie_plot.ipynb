# Pie Plot / Diagram Lingkaran

```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

## Persiapan Data

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08] #semakin besar, semakin jauh
```

## Simple Pie Plot

```python
plt.pie(data, 
        labels=kategori,
        autopct='%1.1f%%', # menamplikan presentase area
        startangle=90)

plt.title('Pie Plot')

plt.show()
```

## Pengaturan Legend dan Color

```python
plt.pie(data, 
        labels=kategori,
        colors=warna,
        autopct='%1.1f%%',
        startangle=90)

plt.legend()
plt.title('Pie Plot')

plt.show()
```

## Exploded Pie Plot

```python
plt.pie(data, 
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Exploded Pie Plot')

plt.show()
```
