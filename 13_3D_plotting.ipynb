```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Persiapan data

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

## 3D Axes

code untuk membuat axes baru (3d)

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

plt.show()
```

## 3D Line Plot

code untuk membuat lineplot baru (3d)

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x, y, z, color='blue')

ax.set(title='3D Line Plot', xlabel='X', ylabel='Y', zlabel='Z')

plt.show()
```

## 3D Scatter Plot

code untuk membuat lineplot baru (3d)

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.scatter(x, y, z, color='green')

ax.set(title='3D Scatter Plot', xlabel='X', ylabel='Y', zlabel='Z')

plt.show()
```
