# Violin Plot

## Cek Versi & Import Modules

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Dataset

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

## Simple Violin Plot

command untuk membuat simple violin plot

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

## Customize Violin Plot

mengatur violin plot

```python
plt.violinplot(data,
               showextrema=True,
               showmeans=False,
               showmedians=False,
               quantiles=[0.25, 0.5, 0.75])

plt.title('Violin Plot')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

## Horisontal Violin Plot

command untuk membuat horizontal violin plot

```python
plt.violinplot(data, vert=False)

plt.title('Horisontal Violin Plot')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

## Multiple Violin Plot

command untuk membuat banyak violin (lebih dari 1 dalam sebuah grafik)

```python
np.random.seed(10)
data1 = np.random.normal(120, 20, 200)
data2 = np.random.normal(70, 30, 200)
data3 = np.random.normal(85, 50, 200)
data4 = np.random.normal(45, 25, 200)

data = [data1, data2, data3, data4]
```

```python
plt.violinplot(data)

plt.title('Multiple Violin Plot')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```
